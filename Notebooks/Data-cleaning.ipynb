```python
# Read house price index into notebook
house_price = pd.read_excel("../Data/raw/average_house_price.xlsx")
house_price.head()
```

```python
house_price.info()
```

```python
#Keep only the columns with borough names i.e drop from column 34 onwards
house_price = house_price.iloc[:,0:34]
house_price.head()
```

```python
house_price=house_price.rename(columns={'Unnamed: 0':'Date'})
house_price.head()
```

```python
#Create month and year column
house_price['Month']= pd.to_datetime(house_price['Date'], format='%m').dt.month_name()
house_price['Year']= pd.to_datetime(house_price['Date'], format='%Y').dt.year
house_price.head()
```

```python
#house_price = house_price.T
house_price = house_price.iloc[1:336,1:36]
```

```python
#house_price = house_price.T
house_price['Year'].value_counts()
```

```python
house_price.index
```

```python
house_price_melted = pd.melt(house_price, id_vars=['Month', 'Year'], value_vars=['City of London', 'Barking & Dagenham', 'Barnet', 'Bexley', 'Brent',
       'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich',
       'Hackney', 'Hammersmith & Fulham', 'Haringey', 'Harrow', 'Havering',
       'Hillingdon', 'Hounslow', 'Islington', 'Kensington & Chelsea',
       'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham',
       'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton',
       'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'Westminster', 'Month',
       'Year'])
```

```python
house_price_melted
```

```python
house_price_melted = house_price_melted.rename(columns={'variable':'Borough','value':'average_house_price'})
house_price_melted
```

```python
house_price_melted.info()
```

```python
#Change data types of year and average house price to round values
```

```python
house_price_melted['Year'] = house_price_melted['Year'].astype(int)
house_price_melted['average_house_price'] = house_price_melted['average_house_price'].astype(float)
house_price_melted['average_house_price'] = house_price_melted['average_house_price'].round(2)
house_price_melted['Year'] = house_price_melted['Year'].round(4)
house_price_melted
```

```python
#Change & in borough name to and 
house_price_melted['Borough'] = house_price_melted['Borough'].replace({'Kensington & Chelsea':'Kensington and Chelsea'
                                                                      , 'Barking & Dagenham' : 'Barking and Dagenham'
                                                                      , 'Hammersmith & Fulham' : 'Hammersmith and Fulham'})
house_price_melted['Borough'].value_counts()
```

```python
#Join data to crime data 
data_joined = pd.merge(data, house_price_melted, left_on=['month_name','year','borough'], 
                       right_on =['Month', 'Year', 'Borough'],  how='left')
```

```python
data_joined
```

```python
data_joined= data_joined.drop(columns=['Month','Year','Borough'])
data_joined
```

```python
data_joined.isnull().sum()
```

```python
#Only looking at crime rates from 2012
data_joined = data_joined[data_joined['year'] >= 2012]
```
